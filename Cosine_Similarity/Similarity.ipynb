```python
def jacardian_distance(document_1_data, document_2_data):
    words_in_doc_1_not_in_doc_2 = len(list(set(document_1_data)-set(document_2_data)))
    words_in_doc_2_not_in_doc_1 = len(list(set(document_2_data)-set(document_1_data)))
    words_in_both_doc_1_and_doc_2 = len(list(set(document_1_data).intersection(document_2_data)))
    try:
        # divide the counts appropiately
        jacardian = words_in_both_doc_1_and_doc_2/(words_in_doc_1_not_in_doc_2+words_in_doc_2_not_in_doc_1)
    except:
        jacardian = 1 
    
    return jacardian


def cosine_similarity(document_1_data, document_2_data):
    #get frequency of words from document 1 & 2
    #get bag of words from document 1&2
    #calculate dot product of two documents vectors (matches with the bag of words)
    #apply normalization for 2 vectors (squaring up the freq and taking sqrt)
    #calculate cosine similarity using the formulae dot prod of 2 vectors / norm_doc_1 * norm_doc_2
    vector1 = Counter(document_1_data)
    vector2 = Counter(document_2_data)
    intersection = set(vector1.keys()) & set(vector2.keys())
    numerator = sum([vector1[x] * vector2[x] for x in intersection])

    sum1 = sum([vector1[x]**2 for x in vector1.keys()])
    sum2 = sum([vector2[x]**2 for x in vector2.keys()])
    denominator = math.sqrt(sum1) * math.sqrt(sum2)
    if not denominator:
        cosine_similarity = 0.0
    else:
        cosine_similarity =  float(numerator) / denominator

    return cosine_similarity
#dot_product_of_two_document_vectors 
# you can refer to the numpy information on how to calculate the dot product of vectors
```
